```python
import sys
import os
import psycopg2

dbname = sys.argv[1]
user = os.environ.get('PGUSER')
password = os.environ.get('PGPASSWORD')
host = os.environ.get('PGHOST')
port = os.environ.get('PGPORT')

conn = psycopg2.connect(
    dbname=dbname,
    user=user,
    password=password,
    host=host,
    port=port
)
cursor = conn.cursor()

x = int(open('./number.txt').read())
year = 2000 + 5*x + 5
query = f'SELECT * FROM book_catalogue WHERE year = {year}'

cursor.execute(query)
for row in cursor.fetchall():
  print(row[0])
```

```python
import os
import sys
import psycopg2

conn = psycopg2.connect(
    dbname=sys.argv[1],
    user=os.environ.get('PGUSER'),
    password=os.environ.get('PGPASSWORD'),
    host=os.environ.get('PGHOST'),
    port=os.environ.get('PGPORT'),
)
cursor = conn.cursor()

date = open('./date.txt').read()
query = f"""
SELECT referees.name FROM matches
JOIN match_referees USING (match_num)
JOIN referees ON match_referees.referee = referees.referee_id
WHERE match_date = '{date}'
 """
cursor.execute(query)
name = cursor.fetchone()[0]

name_words = name.split()
lname = name_words[-1]
rest = [word[0]+'.' for word in name_words[:-1]]
rest = ' '.join(rest)

print(f'{lname} {rest}')
```
